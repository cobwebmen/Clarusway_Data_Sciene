```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
fmri = sns.load_dataset ("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri);
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event", style ="event");
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event", style ="event", markers=True);
```

```python
sns.barplot(x="timepoint", y="signal", data=fmri);
```

```python
sns.boxplot(x="timepoint", y="signal", data=fmri);
```

```python
sns.boxenplot(x="timepoint", y="signal", data=fmri);
```

```python
print(sns.get_dataset_names())
flights = sns.load_dataset("flights")
flights
```

```python
f_sum = flights.groupby(['year','month']).sum()
f_sum
```

```python
sns.lineplot(y= f_sum.passengers, x=f_sum.reset_index().index)
```

```python
flights_wide = flights.pivot(index="year", columns="month", values="passengers")
flights_wide 
```

```python
sns.lineplot(data=flights_wide);
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month");
```

```python

sns.barplot(data=flights, x="year", y="passengers", palette= "vlag")
```

```python
sns.scatterplot(data=flights, x="year", y="passengers", hue= "month" , style="month", legend = "brief")
```

```python
sns.displot(flights.passengers, kde=True, color="green", bins=10)
```

```python
sns.jointplot(data=flights, x="year", y="passengers", color="purple", kind= "reg")
```

```python
sns.jointplot(x="timepoint", y="signal", data=fmri);
```
