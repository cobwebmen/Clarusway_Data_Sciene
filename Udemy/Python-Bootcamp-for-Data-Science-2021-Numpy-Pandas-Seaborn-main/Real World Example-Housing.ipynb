## Real World Example-Housing

First we will import the required libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

Lets now read this dataset into a dataframe using the function **pd.read_csv()**: 

```python
df = pd.read_csv('data/housing.csv')
```

Lets display the head of this dataframe using the function **head()**:

```python
df.head()
```

Now we look at the shape of this dataframe using the attribute **shape**:

```python
df.shape
```

To display more information about this dataset, we use the funcion **info()**:

```python
df.info()
```

To check for missing values, we use the function **isnull()** and the function **sum()**:

```python
df.isnull().sum()
```

We can display the descriptive statistics for the **'sale price'** column using the function **describe()**: 

```python
df['SalePrice'].describe()
```

To create the **correlation matrix** we use the function **cor()**:

```python
df.corr()
```

We can visualize the relationship between the sale price and the four important selected variables using seaborn function **sns.pairplot()**

```python
columns = ['OverallQual', 'TotalBsmtSF', 'GrLivArea', 'GarageArea', 'SalePrice']
sns.pairplot(df[columns])
```

We can create a scatter plot between the variables **'OverallQual'** and **'SalePrice'** using the function **plot.scatter()**:

```python
##
df.plot.scatter(x = 'OverallQual', y = 'SalePrice')
```

We can create a scatter plot between the variables **'TotalBsmtSF'** and **'SalePrice'** using the function **plot.scatter()**:

```python
df.plot.scatter( x = 'TotalBsmtSF', y = 'SalePrice')
```

We can use boolean indexing to display the three **outliers** which are identified in the previous scatter plot like this:

```python
df.loc[(df['TotalBsmtSF'] > 3000) & (df['SalePrice'] < 300000)]
```

Now that we know the index values for these three outliers we can delete them using the function **drop()**:

```python
df = df.drop([332, 523, 1298])
```

After deleting these outliers, lets create the scatter plot again, like this:

```python
df.plot.scatter( x = 'TotalBsmtSF', y = 'SalePrice')
```

Lets check again the **correlations coefficient** between these two variables **after deleting the outliers** using the function **cor()**:

```python
df['TotalBsmtSF'].corr(df['SalePrice'])
```

We can create a scatter plot between the variables **'GrLivArea'** and **'SalePrice'** using the function **plot.scatter()**:

```python
df.plot.scatter( x = 'GrLivArea', y = 'SalePrice')
```

We can create a scatter plot between the variables **'GarageArea'** and **'SalePrice'** using the function **plot.scatter()**:

```python
df.plot.scatter( x = 'GarageArea', y = 'SalePrice')
```

We can use boolean indexing to display the three **outliers** which are identified in the previous scatter plot like this:

```python
df.loc[(df['GarageArea'] > 1200) & (df['SalePrice'] < 300000)]
```

Now that we know the index values for these three outliers we can delete them using the function **drop()**:

```python
df = df.drop([581, 1061, 1190])
```

After deleting these outliers, lets create the scatter plot again, like this:

```python
df.plot.scatter( x = 'GarageArea', y = 'SalePrice')
```

To check for **normality** for the variable **'SalePrice'** we can create a **histogram and a density plot** using the function **histplot()**:

We can create a **density plot** by adding the argument **kde = True**:

```python
sns.histplot(df['SalePrice'], kde = True)
```

Another way to check for normality is to create a probability plot using the **scipy** function **stats.probplot()**:

```python
stats.probplot(df['SalePrice'], plot=plt)
```

To fix any normality violation, we **log transform** these data using numpy function **np.log()**:

```python
df['SalePrice'] = np.log(df['SalePrice'])
```

After transformation, lets check for normality again by creating the histogram, density plot and pair plot, as we have done previously:

```python
sns.histplot(df['SalePrice'], kde = True)
```

```python
stats.probplot(df['SalePrice'], plot=plt)
```

To check for **normality** for the variable **'GrLivArea'** we can create the **histogram, density plot and pair plot**, as we have done in the previous variable:

```python
sns.histplot(df['GrLivArea'], kde = True)
```

```python
stats.probplot(df['GrLivArea'], plot=plt);
```

To make this data normally distributed, we log transform it using the function **np.log()**:

```python
df['GrLivArea'] = np.log(df['GrLivArea'])
```

After transformation, lets check for normality again by creating the histogram, density plot and pair plot:

```python
sns.histplot(df['GrLivArea'], kde = True)
```

```python
stats.probplot(df['GrLivArea'], plot=plt)
```

### Categorical variables

In this dataset we have only one categorical variable which is **kitchen quality**. Lets explore the categories in this variable using the function **value_counts**(), 

```python
df['KitchenQual'].value_counts()
```

To be able to analyze this variable, we need to convert it into **dummy variables**. To convert a categorical variable into dummy variables we use a function in pandas called **get_dummies()**:

```python
kitchen = pd.get_dummies(df['KitchenQual'])
kitchen.head()
```

To add these newly created **dummy variables** to our dataset, we use the function **join()**:

```python
df = df.join(kitchen)
df.head()
```

Because the categorical variable is no longer needed, we can delete it from the dataset using the function **drop()**: 

```python
df = df.drop(['KitchenQual'], axis=1)
df.head()
```
